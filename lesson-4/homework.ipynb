<h3>Task 3<h3>

```python
x = np.linspace(-3, 3, 51)
y1 = x**2
y2 = 2*x + 0.5
y3 = -3*x - 1.5
y4 = np.sin(x)
y_vals = [y1, y2, y3, y4]
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(8, 6))
plt.subplots_adjust(hspace=0.3, wspace=0.3)

ax_1d = ax.flatten()
name = 'График y'

for i, (y, ax) in enumerate(zip(y_vals, ax_1d)):
    ax.plot(x, y)
    ax.set_title(name + str(i+1))
    if i != 0:
        ax.set_xlim(min(x), max(x))

ax_1d[0].set_xticks(np.arange(-5, 7, 2))
plt.show()
```
